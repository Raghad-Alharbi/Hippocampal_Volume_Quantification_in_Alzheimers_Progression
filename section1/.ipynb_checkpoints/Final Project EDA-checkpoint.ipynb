## Loading NIFTI images using NiBabel

NiBabel is a python library for working with neuro-imaging formats (including NIFTI) that we have used in some of the exercises throughout the course. Our volumes and labels are in NIFTI format, so we will use nibabel to load and inspect them.

NiBabel documentation could be found here: https://nipy.org/nibabel/

Our dataset sits in two directories - *images* and *labels*. Each image is represented by a single file (we are fortunate to have our data converted to NIFTI) and has a corresponding label file which is named the same as the image file.

Note that our dataset is "dirty". There are a few images and labels that are not quite right. They should be quite obvious to notice, though. The dataset contains an equal amount of "correct" volumes and corresponding labels, and you don't need to alter values of any samples in order to get the clean dataset.

```python
# TASK: Your data sits in directory /data/TrainingSet.
# Load an image and a segmentation mask into variables called image and label
train_dir = "../data/TrainingSet/"

images_dir = train_dir + "images/"
labels_dir = train_dir + "labels/"

train_images = sorted(glob(images_dir + "*.nii.gz"))
train_labels = sorted(glob(labels_dir + "*.nii.gz"))

print("Number of train images: " , len(train_images))
print("Number of train labels: " , len(train_labels))
```

Note that there is a missing label, which we should discard. 

```python
# Deleting the one image without a label
train_images = [x for x in set(train_images) for y in set(train_labels) if x.split('../data/TrainingSet/images/')[1] == y.split('../data/TrainingSet/labels/')[1]]

print("Number of train images: " , len(train_images))
print("Number of train labels: " , len(train_labels))
```

```python
# Checking if shapes of images and lables are matching
# Deleting the one image without a label
train_images = sorted(train_images)
train_labels = sorted(train_labels)

for image, label in zip(train_images, train_labels):
    image_shape = nib.load(image).shape
    label_shape = nib.load(label).shape
    
    if image_shape != label_shape:
        print(BOLD+"Image and Lable shapes are miss matched in:", image.split("../data/TrainingSet/images/")[1]+END)
        print("Image shape:", image_shape)
        print("label shape:", label_shape)
        train_images.remove(image)
        train_labels.remove(label)
```

```python
print("Number of train images: " , len(train_images))
print("Number of train labels: " , len(train_labels))
```

```python
# Nibabel can present your image data as a Numpy array by calling the method get_fdata()
# The array will contain a multi-dimensional Numpy array with numerical values representing voxel intensities. 
# In our case, images and labels are 3-dimensional, so get_fdata will return a 3-dimensional array. You can verify this
# by accessing the .shape attribute. What are the dimensions of the input arrays?

# TASK: using matplotlib, visualize a few slices from the dataset, along with their labels. 
# You can adjust plot sizes like so if you find them too small:
image_file = nib.load(train_images[1])
label_file = nib.load(train_labels[1])

image = image_file.get_fdata()
label = label_file.get_fdata()

plt.rcParams["figure.figsize"] = (10,30)
fig, axs = plt.subplots(5, 2)

# Axial
num = 0
for i in range(image.shape[0]):
    if (i%8 == 0):
        axs[num, 0].imshow(image[i, :, :], cmap='gray')
        axs[num, 0].set_title('Slice {}'.format(i))
        axs[num, 1].imshow(label[i, :, :], cmap='gray')
        axs[num, 1].set_title('Label for slice {}'. format(i))
        num += 1
```

## Looking at single image data
In this section we will look closer at the NIFTI representation of our volumes. In order to measure the physical volume of hippocampi, we need to understand the relationship between the sizes of our voxels and the physical world.

```python
# Nibabel supports many imaging formats, NIFTI being just one of them. I told you that our images 
# are in NIFTI, but you should confirm if this is indeed the format that we are dealing with
# TASK: using .header_class attribute - what is the format of our images?
## import all images with nibabel:
images = []
labels = []
for image, label in zip(train_images, train_labels):
    images.append(nib.load(image))
    labels.append(nib.load(label))
    
print("Header class for all images :", set([image.header_class for image in images]) )  
print("Header class for all labels :", set([label.header_class for label in labels]) )  
```

```python
# printing a sample header to find the names of the tags
print(images[0].header)
```

Further down we will be inspecting .header attribute that provides access to NIFTI metadata. You can use this resource as a reference for various fields: https://brainder.org/2012/09/23/the-nifti-file-format/

```python
# TASK: How many bits per pixel are used?
bitpix_images = []
bitpix_labels = []
for image, label in zip(images, labels):
    bitpix_images.append(int(image.header['bitpix']))
    bitpix_labels.append(int(label.header['bitpix']))
print("Number of bits per pixel in images: ",Counter(bitpix_images))
print("Number of bits per pixel in labels: ",Counter(bitpix_labels))
```

All images have either 32 bits per pixel or 8 bits per pixel.

- The number of images that has 32 bits per pixel = 205
- The number of images that has 8 bits per pixel = 55

All labels have either 32 bits per pixel or 8 bits per pixel.

- The number of labels that has 32 bits per pixel = 8
- The number of labels that has 8 bits per pixel = 258

```python
# TASK: What are the units of measurement?
spatial_units_images = []
temporal_units_images = []
spatial_units_labels = []
temporal_units_labels = []

for image, label in zip(images, labels):
    spatial_units_images.append(image.header.get_xyzt_units()[0])
    temporal_units_images.append(image.header.get_xyzt_units()[1])
    
    spatial_units_labels.append(label.header.get_xyzt_units()[0])
    temporal_units_labels.append(label.header.get_xyzt_units()[1])

print("Spatial unit for images: ", Counter(spatial_units_images))
print("Temporal unit for images: ", Counter(temporal_units_images))

print("Spatial unit for labels: ", Counter(spatial_units_labels))
print("Temporal unit for labels: ", Counter(temporal_units_labels))
```

The spatial unit of all images and labels is millimeter, and the temporal unit is seconds.

```python
# TASK: Do we have a regular grid? What are grid spacings?
pixdim_images = []
pixdim_labels = []

for image, label in zip(images, labels):
    pixdim_images.append(image.header['pixdim'])
    pixdim_labels.append(label.header['pixdim'])

print("Images grid spacings:", np.unique(np.array(pixdim_images), axis=0) )
print("Labels grid spacings:", np.unique(np.array(pixdim_labels), axis=0) )
```

Yes, we have regular grid, and the grid spacing for x, y, z are all 1 mm. 

```python
# TASK: What dimensions represent axial, sagittal, and coronal slices? How do you know?
dim = images[0].header['dim']
print ("From the 'dim' of the header of the first image:" , images[0].header['dim'] )
print('Sagittal plane: ' , dim [1])
print('Coronal plane: ' , dim [2])
print('Axial plane: ' , dim [3])
```

From the dimentions of the file in the header, the second dimension is the sagittal plane, the third dimension is the coronal plane and the forth dimension is the axial plane. Referring to this nibabel documentation: https://nipy.org/nibabel/coordinate_systems.html

```python
# By now you should have enough information to decide what are dimensions of a single voxel
# TASK: Compute the volume (in mm) of a hippocampus using one of the labels you've loaded. 
# You should get a number between ~2200 and ~4500
labels_fdata = []
for label in labels:
    label_fdata = label.get_fdata()
    labels_fdata.append((label_fdata > 0).sum())
volumes = np.array(labels_fdata)
print( 'Minimum volume: ', min (volumes))
print( 'Maximun volume: ', max (volumes))
```

The volumes of a hippocampus in all labels are between 2382mm and 4401mm, which are in the correct range.

## Plotting some charts

```python
# TASK: Plot a histogram of all volumes that we have in our dataset and see how 
# our dataset measures against a slice of a normal population represented by the chart below.
color_hist = 'teal'
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(volumes, color = color_hist)
fig.suptitle('Distribution of Volumes of a Hippocampus', fontsize = 20)
ax.set_xlabel('Volume in mm')
ax.set_ylabel('Frequency')

plt.show();
```

The distribution looks normal and maybe right skewed a little bit, but with no outliers. Due to removing one image without a label and two images that have shapes not matching their labels.
